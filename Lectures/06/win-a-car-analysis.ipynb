```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

#### Read the data
***

```python
df = pd.read_csv("2020-11-01-win-a-car-data.csv")
df
```

<br>

#### Graphing peoples Stay/Switch choices

***

```python
#We use a catplot to create the bar chart.
sns.catplot(x="stay", kind="count", data=df)
```

True = stay with original choice


<br>

#### Car door random placement
***

```python
#Bar of where the car was placed.
colors = ["red", "green", "blue"]
sns.catplot(x="car", kind="count", palette = colors, data=df)
```

<br>

#### How many people selected different doors?
***

```python
#Bar chart of different picks.
colors1 = ["blue", "red", "green"]
sns.catplot(x="pick", kind="count", palette = colors1, data=df)
```

<br>

#### Is changing a good idea?
***

```python
#is car value the same as pcik value and is the answer the sane as df["stay"]?
#on the right: was car different to pick? and did they not stay?
#this statement returns True if the person wins the car.
df["win"] = ((df["car"] == df["pick"]) & df["stay"]) | ((df["car"] != df["pick"]) & ~df["stay"]) 
```

```python
df
```

```python
df["win"].sum() #checking count of wins
```

```python
#Bar chart of winning vs losing. Exactly half (coincidentally)
sns.catplot(x="win", kind="count", data=df)
```

```python
#Bar chart of winning vs losing.
sns.catplot(x="stay",hue="win", kind="count", data=df)
```

If you stay you are much more likely to lose. If you dont stay you are much more likely to win.
